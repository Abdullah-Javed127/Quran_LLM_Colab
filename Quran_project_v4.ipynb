```python
from datasets import load_dataset

ds = load_dataset("M-AI-C/quran-en-tafssirs")
```

```python
from datasets import DatasetDict

my_dataset=DatasetDict({"train":ds["train"].remove_columns(['en-ahmedraza', 'en-arberry', 'en-asad', 'en-daryabadi', 'en-hilali',
                                                            'en-itani', 'en-maududi', 'en-mubarakpuri', 'en-pickthall', 'en-qarai', 'en-qaribullah', 'en-sahih',
                                                            'en-sarwar', 'en-shakir', 'en-transliterati', 'en-wahiduddi', 'en-yusufali',
                                                            'en-tafsir-mokhtasar-text', 'en-tafsir-maarif-html', 'en-tafsir-maarif-text', 'en-tafsir-ibn-kathir-html',
                                                            'en-tafsir-ibn-kathir-text'])})
```

```python
# Define a function to combine the fields into a single text string.
def combine_fields(example):
    combined = (
        f"Translation: {example['en-ahmedali']}\n"
        f"Ayah: {example['ayah']}\n"
        f"Surah: {example['sorah']}\n"
        f"Arabic: {example['sentence']}\n"
        f"Tafsir: {example['en-tafsir-mokhtasar-html']}"
    )
    example["combined_text"] = combined
    return example

# Apply the function to every entry in the train split.
my_dataset = my_dataset.map(combine_fields)
```

```python
# -----------------------------------------
# 2. Generate Embeddings for the Dataset
# -----------------------------------------
# Load a lightweight embedding model.
embedder = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings for the combined text of each entry.
texts = my_dataset['train']['combined_text'] # Accessing the 'train' split, then the 'combined_text' column
embeddings = embedder.encode(texts, convert_to_tensor=False)
```

```python

# -----------------------------------------
# 3. Build the FAISS Index for Retrieval
# -----------------------------------------
# Convert embeddings to a NumPy array (FAISS requires float32 arrays).
embeddings_np = np.array(embeddings).astype('float32')
dimension = embeddings_np.shape[1]  # e.g., 384 for all-MiniLM-L6-v2

# Create a FAISS index using L2 distance.
index = faiss.IndexFlatL2(dimension)
index.add(embeddings_np)
print(f"FAISS index built with {index.ntotal} vectors.")
```
